# 1. 오차 역전파의 개념

- 가중치는 오차 역전파 방법을 사용해 수정함
- 경사 하강법의 확장 개념
- 임의의 가중치를 선언하고 결괏값을 이용해 오차를 구한 뒤 이 오차가 최소인 지점으로 계속해서 조금씩 이동시킴
- 오차 역전파(back propagation):
  - 다층 퍼셉트론에서의 최적화 과정
- 오차 역전파 구동 방식
  1. 임의의 초기 가중치(w)를 준 뒤 결과(y)를 계산함
  2. 계산 결과와 우리가 원하는 값 사이의 오차를 구함
  3. 경사 하강법을 이용해 바로 앞 가중치를 오차가 작아지는 방향으로 업데이트함
  4. 위 과정을 더이상 오차가 줄어들지 않을 때까지 반복함
