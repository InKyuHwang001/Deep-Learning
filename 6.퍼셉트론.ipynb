- 신경망을 이루는 가장 중요한 기본 단위는 퍼셉트론(perceptron)



# 1.가중치, 가중합, 바이어스, 활성화 함수

- `y=wx+b`: 
  - w는 가중치 b는 바이어스
  - 가중합: w*x의 합
  - 활성화 합수: 가중합의 결과를 놓고 1 또는 0을 출력해서 다음으로 보냄, 여기서 0과 1을 판단하는 함수

# 2.퍼셉트론의 과제

- 여러 개의 선을 아무리 그어보아도 하나의 직선으로는 흰점과 검은점을 구분
할 수 없음
- 퍼셉트론 역시 선을 긋는 작업이라고 할 수 있음
- 선을 아무리 그어도 해결되지 않는 상황이 있음

# 3.XOR 문제


- XOR 문제: 논리 회로에 등장하는 개념
  - 둘 중 하나만이 이면 결괏값이 1로 출력
- 게이트(gate): 0과 1을 입력해 하나의 값을 출력하는 회로가 모여 만들어지는 회로

- AND와 OR 게이트는 직선을 그어 결괏값이 1인 값을 구별
- XOR의 경우 선을 그어 구분할 수 없음

- 해결한 개념이 바로 다층퍼셉트론(multilayer perceptron)
