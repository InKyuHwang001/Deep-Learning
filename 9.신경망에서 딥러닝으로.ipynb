# 1.기울기 소실 문제와 활성화 함수

- 기울기 소실(vanishing gradient) 문제가 발생하기 시작한 것은 활성화 함수로 사용된 시그모이드 함수의 특성 때문
  - 여러 층을 지나면 기울기가 사라져 가중치를 수정하기 어려움
  - 활성화 함수를 시그모이드 함수 말고 쓰기 시작해 해결
- 활성화 함수(시그모이드 제외)
  1. 하이퍼볼릭 탄젠트 함수
    - 미분한 값의 범위가 함께 확장되는 효과를 가져옴
    - 여전히 1보다 작은 값이 존재하므로 기울기 소실 문제는 사라지지 않음
  2. 렐루 함수
    - 가장 많이 사용되는 활성화 함수임
  3. 소프트플러스 함수
    - 이후 렐루의 0이 되는 순간을 완화


# 2.속도와 정확도 문제를 해결하는 고급 경사 하강법


1. 확률적 경사 하강법
  - 경사 하강법은 속도가 느림
  - 전체 데이터가 아니라 랜덤하게 추출한 일부 데이터 사용
    - 더 빨리 업데이트됨
2. 모멘텀
  - 모멘텀 SGD란 경사 하강법에 탄력을 더해 주는 것
  - 수정 방향이 양수방향으로, 음수방향으로 지그재그로 일어나는 현상이 줄어들고, 이전 이동 값을 고려하여 일정 비율만큼만 효과를 냄
